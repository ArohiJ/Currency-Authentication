```python
import cv2
import numpy as np
from scipy.stats import *
from scipy import ndimage
from tkinter import *
from PIL import ImageTk, Image
import os
import pandas as pd
import tkinter as tk
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg,NavigationToolbar2Tk 
from datascience import *
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import *
```

```python
def data_analyse():
    
    data = "C:\\Users\\arohi_000\\Desktop\\College stuff\\Mini Project\\Mini Project sem-5\\" 
    banknotes = Table.read_table(data + 'BankNote_Authentication.csv') #reading table from path
    
    color_table = Table().with_columns(
    'class', make_array(1, 0),
    'Color', make_array('darkblue', 'gold')
    )
    
    banknotes = banknotes.join('class', color_table) 
    return banknotes
```

```python
def data_mdl():
    
    win5=Tk()
    
    banknotes=data_analyse()

    figure = plt.Figure(figsize=(10,10))
    ax = figure.add_subplot(111, projection='3d')
    #px.scatter_3d(banknotes, x=banknotes.column('skewness'), y= banknotes.column('variance'), z=banknotes.column('curtosis'),
                    #color=banknotes.column('Color'))

    ax.scatter(banknotes.column('skewness'),banknotes.column('variance'),banknotes.column('curtosis'),
               c=banknotes.column('Color'))
    scatter1 = FigureCanvasTkAgg(figure, win5) 
     
    ax.set_xlabel('Skewness')
    ax.set_ylabel('Variance')
    ax.set_zlabel('Kurtosis')
    ax.set_title('Skewness, Variance and Kurtosis Graph')
    
    toolbar=NavigationToolbar2Tk(scatter1, win5)
    toolbar.update()
    toolbar.pack()
    scatter1._tkcanvas.pack()
    scatter1.get_tk_widget().pack(side = BOTTOM, fill = BOTH)  
    
    win5.title('Data Set Analyser')
    win5.configure(background='dark olive green')
    win5.geometry("900x500+100+50")
    win5.mainloop()
```

```python
def knn_mdl():
    
    win4=Tk()
    
    banknotes=data_analyse()
    
    y=np.asarray(banknotes.select('class').rows)
    x=np.asarray(banknotes.select('variance','skewness','curtosis').rows)
    
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
    
    knn = KNeighborsClassifier(n_neighbors=31)  
    knn.fit(x_train, np.ravel(y_train))
    
    knn.predict(x_test)
    accuracy = str(knn.score(x_test, y_test))
    
    test_class=Label(win4, text="Data set has been split to 75% training data and 25% testing data.", fg='black', font=("Helvetica", 18))
    test_class.place(x=100, y=150)
    test_class.configure(background='dark olive green')
    
    test_knn=Label(win4, text="The data model has been built based on 31 nearest neighbours.", fg='black', font=("Helvetica", 18))
    test_knn.place(x=100, y=225)
    test_knn.configure(background='dark olive green')
    
    acc=Label(win4, text="Accuracy of model:"+accuracy, fg='black', font=("Helvetica", 18))
    acc.place(x=100, y=300)
    acc.configure(background='dark olive green')
    
    win4.title('KNN Analyser')
    win4.configure(background='dark olive green')
    win4.geometry("900x500+100+50")
    win4.mainloop()   
```

```python
def img_cal(path):
    
    win3=Tk()
    
    im = Image.open(path)
    c_im = im.crop((0,0,400,400))
    ph = ImageTk.PhotoImage(c_im, master=win3)
    label = Label(win3, image=ph)
    label.image=ph
    label.place(x=700, y=100)
    
    d_img = cv2.imread(path, 0)
    crop_img = d_img[0:400, 0:400]
    
    v=str(variation(crop_img, axis=None))
    k=str(kurtosis(crop_img, axis=None))
    s=str(skew(crop_img, axis=None))
    e=str(entropy(crop_img, axis =None))
    
    var=Label(win3, text="Variance of the image:"+v, fg='black', font=("Helvetica", 18))
    var.place(x=100, y=150)
    var.configure(background='dark olive green')
    
    kurt=Label(win3, text="Kurtosis of the image:"+k, fg='black', font=("Helvetica", 18))
    kurt.place(x=100, y=200)
    kurt.configure(background='dark olive green')
    
    sk=Label(win3, text="Skewness of the image:"+s, fg='black', font=("Helvetica", 18))
    sk.place(x=100, y=250)
    sk.configure(background='dark olive green')
    
    ent=Label(win3, text="Entropy of the image:"+e, fg='black', font=("Helvetica", 18))
    ent.place(x=100, y=300)
    ent.configure(background='dark olive green')
    
    x_test=np.asarray([[float(v),float(s),float(k)]])
    
    def img_mdl():
        
        banknotes = data_analyse()
        
        y_train=np.asarray(banknotes.select('class').rows)
        x_train=np.asarray(banknotes.select('variance','skewness','curtosis').rows)
        knn = KNeighborsClassifier(n_neighbors=31)  
        knn.fit(x_train, np.ravel(y_train))
        ans = knn.predict(x_test)
        
        if(ans==[1]):
            ans='real !!!'
        else:
            ans='couterfeit !!!'
        
        ans_print=Label(win3, text="The given currency is "+ans, fg='ivory', font=("Helvetica", 18))
        ans_print.place(x=100, y=530)
        ans_print.configure(background='dark olive green')
    
    btn=Button(win3, text="Predict Class", fg='black', command=img_mdl)
    btn.place(x=200, y=400)
    btn.config(height = 5, width = 50, background='ivory')
    
    win3.title('Data from Image')
    win3.configure(background='dark olive green')
    win3.geometry("1300x800")
    win3.mainloop()    
```
